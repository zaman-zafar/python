# Importing and Cleaning Data

```python
gdpdata = pd.read_csv(GDPData)
country = pd.read_csv(Country)
```

```python
country_ = country.copy()
country_["alpha3"] = country_["alpha3"].str.upper()
country_.shape
```

```python
gdpdata.columns
```

```python
gdpdata_ = gdpdata.drop(columns=["2020" , "Indicator Name" , "Indicator Code"] , axis=1)
```

```python
gdpdata_.shape
```

```python
gdpdata_ = gdpdata_.dropna(axis=0, thresh=30)
```

```python
gdpdata_.shape
```

```python
gdpdata_.head()
```

```python
data = gdpdata_.copy()
data["avg"] = data.mean(axis =1)
```

```python
data.shape
```

## Merging DataFrames

```python
k = pd.merge(data , country_ , left_on= "Country Code" , right_on= "alpha3")
```

```python
fin = k.copy()
fin = k.nlargest(15 , "avg")
```

```python
fina = fin.dropna(axis = 'columns' ,  thresh= 2 , how='all' )
```

```python
fina.columns
```

```python
final = fina.drop(columns = ['Country Code' , 'avg' , 'id' , 'name' , 'alpha2' , 'alpha3' ], axis = 1)
```

## Transposing and creating Date Column

```python
finals = final.set_index("Country Name").T

finals.index = pd.to_datetime(finals.index)
finals.head()
```

# Converting data in Billions USD

```python
finals = finals.divide(1000000000)
```

```python
finals.head()
```

## Creating Bar Chart Race

```python
def summary(values, ranks):
    total = int(round(values.sum(), -2))
    s = f'In Billions USD'
    return {'x': .99, 'y': .99, 's': s, 'ha': 'right', 'size': 8}
```
