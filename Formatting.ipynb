```python
import pandas as pd
import numpy as np
```

```python
data = np.random.uniform(-2 , 2 , size = (6,8))
df = pd.DataFrame(data )
df

# https://www.codegrepper.com/code-examples/python/python+fstring
```

# Using f-strings to highlight negative values

```python
def negative_color(x):
    if x > 0: 
        color = 'black'
    else: 
        color = 'red'
    
    return f'color: { color }'
```

```python
df.style.applymap(negative_color)
```

# To highlight the biggest value in each column

```python
colmax = df.style.highlight_max(color="#CBCE91FF")
colmax
```

# To highlight the biggest value in entire Dataframe

```python
df.style.highlight_max( axis=None , color="#CBCE91FF")
```

# To highlight biggest value in each row

```python
rowmax = df.style.highlight_max( axis=1 , color="#CBCE91FF")
rowmax
```

# Adding Bar plot to pandas DataFrame

```python
df.style.bar(width=90 , color= [ "#EEA47FFF" , "#9CC3D5FF" ], align= "mid" , vmin=(-2) , vmax = (2))
```

# To change color while hovering

```python
df.style.set_table_styles([{'selector': 'tr:hover','props': [('background-color', 'pink')]}])
```
