# Creating Pivot Table for category and subcategory

```python
salesbycategory = pd.pivot_table(finaldata , columns= "Category" , index=["Subcategory"] 
                                 , values = 'Net Price' ,fill_value= "", aggfunc=np.sum , margins=True)
```

```python
salesbycategory
```

# Pivot table for sales by category by year

```python
salesbyyear = pd.pivot_table(finaldata , columns= pd.DatetimeIndex(salestable['Order Date']).year , index=["Category"] , values = 'Net Price' ,fill_value= "", aggfunc=np.sum )
```

```python
salesbyyear
```

# Using iloc to slice data for 2017 from pivot table

```python
data2017 = pd.DataFrame(salesbyyear.iloc[:,0])
data2017
```
